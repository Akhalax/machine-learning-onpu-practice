### 1. Создать новый столбец, который будет содержать информацию о количестве только данных слов на каждый комментарий (а не всех слов на комментарий, как в примере). 
#### Новый столбец - словарь. При выполнении данного пункта можно использовать функцию graphlab.SFrame.apply. 
##### products['awesome'] = products['word_count'].apply(awesome_count) 

```python
product['word_count'] = graphlab.text_analytics.count_words(product['text'])
product.head()
```

```python
has_key = product[product['word_count'].dict_has_any_keys(selected_words)]
has_key
```

```python
def right_words_count(d):
    awesome = d.get('awesome',0)
    great = d.get('great',0)
    fantastic = d.get('fantastic',0) 
    amazing = d.get('amazing',0)
    love = d.get('love',0)
    horrible = d.get('horrible',0)
    bad = d.get('bad',0)
    terrible = d.get('terrible',0)
    awful = d.get('awful',0)
    wow = d.get('wow',0)
    hate = d.get('hate',0)
    
    return {'awesome':awesome, 'great':great, 'fantastic':fantastic, 'amazing':amazing, 'love':love, 
            'horrible':horrible, 'bad':bad,'terrible':terrible, 'awful':awful, 'wow':wow, 'hate':hate}

has_key['right_words'] = has_key['word_count'].apply(right_words_count)
has_key
```

### 2. Создать новую модель, используя в качестве фичи новый список слов из пункта 1.

```python
has_key = has_key[has_key['score'] != 3]
has_key['sentiment'] = has_key['score'] >=4

train_data,test_data = has_key.random_split(.8, seed=0)
```

```python
sentiment_model = graphlab.logistic_classifier.create(train_data,
                                                     target='sentiment',
                                                     features=['right_words'],
                                                     validation_set=test_data)
```

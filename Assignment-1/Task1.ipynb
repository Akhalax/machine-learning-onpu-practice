### Генерация двух одномерных массивов, заполненных случайными значениями в диапазоне [-20,20]

```python
import numpy as np
n = 30
X = np.random.randint(-20, 21, size=(n))
Y = np.random.randint(-20, 21, size=(n))
#X = np.array([  2,   0,  -2,  11,  -9,  18,   9,   7,  18, -13])
#Y = np.array([ 18, -13,  14,  -8,  -1, -10,  12, -12, -14, -13])
print "Array X:"
print X
print "Array Y:"
print Y
```


### Отрисовка полученных координат 

```python
import matplotlib.pyplot as plt
#plt.figure()
plt.title('X array')
plt.plot(X, Y, color="g")
plt.show()
```

### Сортировка массива Х выбором

```python
def selectionSort(array):
    for item in range(len(array) - 1, 0, -1):
        max_value_position = 0
        for location in range(1, item + 1):
            if array[location] > array[max_value_position]:
                max_value_position = location

        array[item], array[max_value_position] = \
            array[max_value_position], array[item]
```

```python
selectionSort(X)
print 'Sorted array X: '
print X
```

### Отрисовка полученных координат в виде точек красного цвета

```python
plt.title('Sorted array X')
plt.plot(X, Y, 'ro')
plt.show()
```
